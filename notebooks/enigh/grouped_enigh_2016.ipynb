```python
file = "../../dataframes/enigh/enigh_transf_2016.csv"
```

```python
df = pd.read_csv(filepath_or_buffer=file)
```

```python
df_copy = df.copy()
```

```python
df_copy.drop(labels=['Unnamed: 0'], axis=1, inplace=True)
```

```python
df_copy.columns
```

```python
df_copy.shape
```

```python
categoricos = ['tam_loc', 'est_socio', 'clase_hog', 'sexo_jefe', 'educa_jefe']
```

```python
numericos = []

for row in df_copy.columns:
    if row not in categoricos:
        if (row != 'ent') & (row != 'mun') & (row != 'loc'):
#            print(row)
            numericos.append(row)
            
numericos
```

```python
f = {}

for column in df_copy.columns:
#    print(column)
    if(column in df_copy[categoricos].columns):
#        print(column, 'categoricos')
        f[column] = lambda x: x.value_counts().index[0]
    elif(column in df_copy[numericos].columns):
#        print(column, 'numericos')
        f[column] = 'mean'
    
f
```

```python
grouped = df_copy.groupby(by=['ent', 'mun']).agg(f)
```

```python
grouped.head()
```

```python
grouped.educa_jefe.value_counts()
```

```python
file = '../../dataframes/enigh/enigh_grouped_2016.csv'
```
