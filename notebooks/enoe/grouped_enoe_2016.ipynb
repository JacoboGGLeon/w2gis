```python
file = "../../dataframes/enoe/enoe_transf_2016.csv"
```

```python
df = pd.read_csv(filepath_or_buffer=file)
```

```python
df_copy = df.copy()
```

```python
df_copy.drop(labels=['Unnamed: 0'], axis=1, inplace=True)
```

```python
df_copy.columns
```

```python
numericos = ['eda', 'nac_dia', 'nac_mes', 'nac_anio', 'anios_esc', 'hrsocup', 
             'ingocup', 'ing_x_hrs', 't_tra']
```

```python
categoricos = []

for row in df_copy.columns:
    if row not in numericos:
        if (row != 'ent') & (row != 'mun') & (row != 'loc'):
#            print(row)
            categoricos.append(row)
            
categoricos
```

```python
df_copy.shape
```

```python
f = {}

for column in df_copy.columns:
#    print(column)
    if(column in df_copy[categoricos].columns):
#        print(column, 'categoricos')
        f[column] = lambda x: x.value_counts().index[0]
    elif(column in df_copy[numericos].columns):
#        print(column, 'numericos')
        f[column] = 'mean'
    
f
```

```python
grouped = df_copy.groupby(by=['ent', 'mun', 'loc']).agg(f)
```

```python
grouped
```

```python
file = '../../dataframes/enoe/enoe_grouped_2016.csv'
```

```python
grouped.to_csv(file)
```
